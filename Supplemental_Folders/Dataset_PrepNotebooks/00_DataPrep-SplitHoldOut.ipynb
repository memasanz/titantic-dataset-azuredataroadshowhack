```python
import pandas as pd
```

```python
df = pd.read_csv('./Data/Hold.csv')
df = df.sample(frac=8/10).reset_index(drop=True)
df.shape
```

```python
df.columns
```

```python
columns_train1 = ['passenger_id','fare', 'cabin', 'embarked','home.dest', 'survived'] 

columns_train2 = ['passenger_id', 'pclass', 'name', 'sex', 'age', 'sibsp', 'parch', 'ticket'] 

df_train1 = df[columns_train1]
df_train2 = df[columns_train2]

df_train1.to_csv('./Data/Day2_Inference1.csv', index = False)
df_train2.to_csv('./Data/Day2_inference2.csv', index = False)
```
