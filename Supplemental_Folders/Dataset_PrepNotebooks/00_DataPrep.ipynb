## Data Preperation

- https://www.kaggle.com/vinicius150987/titanic3

```python
import pandas as pd
```

```python
df = pd.read_csv('./Data/full_titantic.csv')
df = df.sample(frac=1).reset_index(drop=True)
df.to_csv('./train/data.csv', index = False)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('survived', axis = 1)
Y = df['survived']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)


df_y_train =  y_train.to_frame(name='survived')
df_y_test =   y_test.to_frame(name ='survived')
```

```python
# Train Dataset

X_train.reset_index(inplace=True)
X_train = X_train.rename(columns = {'index':'passenger_id'})

df_y_train.reset_index(inplace=True)
df_y_train = df_y_train.rename(columns = {'index':'passenger_id'})

df_train = X_train.copy()
df_train['survived'] = df_y_train['survived']

print(df_train.shape)
print(df_train.columns)

columns_train1 = ['passenger_id','fare', 'cabin', 'embarked','home.dest', 'survived'] 

columns_train2 = ['passenger_id', 'pclass', 'name', 'sex', 'age', 'sibsp', 'parch', 'ticket'] 

df_train1 = df_train[columns_train1]
df_train2 = df_train[columns_train2]

df_train1.to_csv('./Data/Train1.csv', index = False)
df_train2.to_csv('./Data/Train2.csv', index = False)
```

```python
# Holdout Dataset

X_test.reset_index(inplace=True)
X_test = X_test.rename(columns = {'index':'passenger_id'})
print(df_test.shape)
df_y_test.reset_index(inplace=True)
df_y_test = df_y_test.rename(columns = {'index':'passenger_id'})

df_test = X_test.copy()
print(df_test.shape)
df_test['survived'] = df_y_test['survived']

print(df_test.shape)
df_test.drop(["body", "boat"], axis='columns', inplace = True)
print(df_test.columns)
df_test.to_csv('./Data/Hold.csv', index = False)
```
